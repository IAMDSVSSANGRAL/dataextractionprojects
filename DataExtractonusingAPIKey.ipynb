```python
# Initialize an empty DataFrame
movie_df = pd.DataFrame()

for i in range(1, 467):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=9b175b42e259140e5dedd9fad7499daa&language=en-US&page={}'.format(i))
    new_df = pd.DataFrame(response.json()["results"])[['id', 'original_language', 'original_title', 'overview', 'popularity', 'release_date', 'title', 'vote_average', 'vote_count']]
    #movie_df = movie_df.append(new_df, ignore_index=True)
    movie_df = pd.concat([movie_df, new_df], ignore_index=True)

    # print(response)
    # print(new_df)
```
