```python
import requests

url = "https://covid-193.p.rapidapi.com/statistics"

headers = {
	"X-RapidAPI-Key": "232932f522msh53e45bb95dc46a9p141d94jsn122ac3772ff2",
	"X-RapidAPI-Host": "covid-193.p.rapidapi.com"
}

response = requests.get(url, headers=headers)
print(response.json())
```

```python
response.json()['get']
```

```python
response.json()['results']
```

```python
df = pd.DataFrame(response.json()['response'])
```

```python
df.tail()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
# Drop rows with null values
df_cleaned = df.dropna()
```

```python
df_cleaned.shape
```

```python
expanded_df_cases = pd.json_normalize(df_cleaned['cases'])
```

```python
expanded_df_cases
```

```python
expanded_df_cases.isna().sum()
```

```python
expanded_df_cases.shape
```

```python
expanded_df_deaths = pd.json_normalize(df_cleaned['deaths'])
```

```python
expanded_df_deaths
```

```python
expanded_df_deaths.isna().sum()
```

```python
expanded_df_tests = pd.json_normalize(df_cleaned['tests'])
```

```python
expanded_df_tests
```

```python
expanded_df_tests.isna().sum()
```

```python
# Concatenate along columns
result = pd.concat([df_cleaned, expanded_df_cases, expanded_df_deaths, expanded_df_tests], axis=1)
```
